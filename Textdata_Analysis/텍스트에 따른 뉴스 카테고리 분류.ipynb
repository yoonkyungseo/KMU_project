```python
# 한국어 불용어 처리 참고
ko_Stopword = list(pd.read_csv('한국어 불용어.csv').불용어)
```

```python
stanza.download('ko')
nlp = stanza.Pipeline('ko')
```

```python
def extract_nouns_stanza(text):
    doc = nlp(text)
    for sentence in doc.sentences:
        for word in sentence.words:
            lemma = word.lemma.split('+')
            xpos = word.xpos.split('+')
            for lem, pos in zip(lemma, xpos):
                if pos.startswith('n'):
                    yield lem
```

```python
cv = CountVectorizer(max_features=100, tokenizer=extract_nouns_stanza, stop_words=ko_Stopword)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
```

--------------------

```python
# optimizer = 어떻게 학습시킬 것인가, adam = 변형된 경사하강법
# loss = 손실함수, binary~ = 교차엔트로피, matrics = 평가지표(예측한 값을 확인하기 위해 설정)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train.A,y_train,epochs=5, batch_size=20, callbacks=[es,mc])
```

```python
model.evaluate(x_test.A, y_test)
```

### 기사 주제분석

```python
for i in range(len(df)):
    df['content'][i] = BeautifulSoup(df["content"][i]).get_text()

for i in range(len(df)):
    df['content'][i] = re.sub("[^ㄱ-ㅣ가-힣]",           # 바꿀패턴:한글이 아닌 모든 것 - 영어도 없애는 것이 맞나?
                      " ",                   # 바뀐내용:공백으로 바꿔라
                      df['content'][i] )  # 바꿀문서
```

```python
token_re = re.compile(r'\b\w{2,}\b', re.UNICODE)
def tokenizer(text):
    text = text.lower()
    words = []
    for word in token_re.findall(text):
        if word not in ko_Stopword:
            words.append(word)
    return words

docs = []
for text in df.content:
    doc = tokenizer(text)
    docs.append(doc)

dic = Dictionary(docs)

corpus = []
for doc in docs:
    bow = dic.doc2bow(doc)
    corpus.append(bow)

dic.filter_extremes(no_below=10, no_above=0.7)

corpus = []
for doc in docs:
    bow = dic.doc2bow(doc)
    corpus.append(bow)
```

```python
train_corpus, valid_corpus = train_test_split(corpus, test_size=0.1, random_state=0)
```

```python
model = LdaModel(
    corpus=train_corpus,
    id2word=dic,
    num_topics=100,
    random_state=0)
```

```python
old_loss = -numpy.inf
loss = model.log_perplexity(valid_corpus)
loss
```

```python
while loss > old_loss + 0.1:
    model.update(train_corpus)
    old_loss = loss
    loss = model.log_perplexity(valid_corpus)
    print(loss)
```

```python
model.save('lda-model')
```
